```python
from groq import Groq
from google.colab import userdata

api = userdata.get('API_GROQ')
client = Groq(api_key=api)

messages = [
    {
        "role": "system",
        "content": "Kamu adalah asisten yang bertanggung jawab untuk menghitung dan menyajikan nilai rapor kumulatif siswa berdasarkan data yang diberikan oleh pengguna."
    }
]

data_nilai = list(map(int, input("Masukkan nilai peserta didik: ").split()))
nilai_bonus = []
count = len(data_nilai)

if count > 6:
  for i in range(6, count):
    nilai_bonus.append(int(data_nilai[i]))

summary = f"""
Total data: {count}
Nilai ulangan harian 1: {data_nilai[0]}
Nilai ulangan harian 2: {data_nilai[1]}
Nilai ulangan harian 3: {data_nilai[2]}
Nilai ulangan tengah semester: {data_nilai[3]}
Nilai ulangan akhir semester: {data_nilai[4]}
Nilai sikap: {data_nilai[5]}
Nilai bonus: {nilai_bonus}
"""

messages.append({"role": "user", "content": f"Berdasarkan data berikut: {summary}, tentukan nilai akhir siswa dalam rentang 0–100 secara jelas dan menggunakan bahasa formal. Nilai bonus dapat dipertimbangkan sebagai tambahan, tetapi tidak menjadi faktor utama dalam penilaian."})
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=messages
)

bot_reply = response.choices[0].message.content
print(f"Komentar bot: {bot_reply}")
```
