```python
from groq import Groq
from google.colab import userdata

api = userdata.get('API_GROQ')
client = Groq(api_key=api)

messages = [
    {
        "role": "system",
        "content": "Kamu adalah asisten yang memberikan ringkasan hasil validasi data dengan gaya singkat dan jelas."
    }
]

data_nilai = input("Masukkan nilai peserta didik: ").split()
nilai_real = []
count = 0

print("List data yang tidak valid:")

for nilai in data_nilai:
  if isinstance(nilai, int):
    nilai_real.append(int(nilai))
  elif isinstance(nilai, str):
    if nilai.isdigit():
      nilai_real.append(int(nilai))
    else:
      count+=1
      print(f"{count}. Nilai ({nilai}) tidak valid.")
  else:
    count+=1
    print(f"{count}. Nilai ({nilai}) tidak valid karena bertipe data ({type(nilai).__name__})")

if count == 0:
  print("Semua data valid.")

summary = f"""
Total data: {len(data_nilai)}
Data yang valid: {len(nilai_real)}
Data yang tidak valid: {count}
Nilai bersih: {nilai_real}
"""

messages.append({"role": "user", "content": f"Buatkan ringkasan hasil dalam 1 kalimat dari data: {summary}, jelaskan secara ilmiah."})

response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=messages
)

bot_reply = response.choices[0].message.content
print(f"Komentar bot: {bot_reply}")
```
